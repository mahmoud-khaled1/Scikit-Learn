### 2.1-SelectPrecentile 

Select features according to a percentile of the highest scores.

<img src="Chi2.png">

```python
from sklearn.feature_selection import SelectPercentile 
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import chi2,f_classif # what is type of calculation 
```

```python
Breast_Cancer=load_breast_cancer()
X_train=Breast_Cancer.data
Y_train=Breast_Cancer.target
```

```python
print("X_Train Shape : ",X_train.shape)
print("Y_train Shape : ",Y_train.shape)
```

```python
print("Feature Name : \n",Breast_Cancer.feature_names)
```

```python
Feature_Selection=SelectPercentile(score_func=chi2,percentile=20)#take all the feature and return 20% of features 
```

```python
X_train=Feature_Selection.fit_transform(X_train,Y_train)
```

```python
print("Feature Selection : ",Feature_Selection.get_support())
```

```python
print("X Train after Feature Selection : ",X_train.shape)
```
